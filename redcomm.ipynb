```python
# CONVERT date's object to datetime
redcomm['date'] = pd.to_datetime(redcomm['date'], errors='coerce')
```

MENGUBAH SEJUMLAH OBJECT MENJADI NUMERIC

```python
# CHANGE OBJECTS to NUMERIC

redcomm['like'] = pd.to_numeric(redcomm['like'].str.replace(',',''), errors='coerce')
```

```python
redcomm.dtypes
```

```python
redcomm
```

```python
redcomm['comments'] = pd.to_numeric(redcomm['comments'], errors='coerce')
redcomm['saved'] = pd.to_numeric(redcomm['saved'], errors='coerce')
redcomm['engagement'] = pd.to_numeric(redcomm['engagement'].str.replace(',',''), errors='coerce')
redcomm['video_views'] = pd.to_numeric(redcomm['video_views'].str.replace(',',''), errors='coerce')
redcomm['reach'] = pd.to_numeric(redcomm['reach'].str.replace(',',''), errors='coerce')
redcomm['impression'] = pd.to_numeric(redcomm['impression'].str.replace(',',''), errors='coerce')
```

```python
redcomm.dtypes
```

```python
redcomm
```

SET KOLOM DAY DAN MONTH

```python
redcomm['day'] = redcomm['date'].dt.day_name()
redcomm['month'] = redcomm['date'].dt.month
```

```python
redcomm
```

SET KOLOM TIME UNTUK SETUP INTERVAL PERIODE HARI: Midnight, Morning, Afternoon, Evening, Night

```python
redcomm['time']=pd.to_datetime(redcomm['date'], errors='coerce').dt.time
```

```python
redcomm
```

```python
redcomm.dtypes
```

```python
redcomm['hour']= pd.to_datetime(redcomm['date']).dt.hour
```

```python
redcomm.dtypes
```

```python
redcomm
```

PEMBAGIAN INTERVAL WAKTU

```python
# PEMBAGIAN INTERVAL WAKTU

interval_hour = [0, 6, 12, 18, 24]
labels = ['Midnight', 'Morning', 'Afternoon', 'Evening']

redcomm['period'] = pd.cut(redcomm['hour'], bins=interval_hour, labels=labels, include_lowest=True, right=False)
```

```python
redcomm
```

KOLOM TAHUN

```python
redcomm['year']= pd.to_datetime(redcomm['date']).dt.year
redcomm
```

ENGAGEMENT HARIAN

```python
redcomm_daily = redcomm[['day', 'like', 'comments', 'saved', 'share', 'engagement', 'video_views', 'reach', 'impression']]
redcomm_daily
```

```python
engagedaily = pd.DataFrame(redcomm_daily.groupby('day')[['engagement', 'reach', 'impression', 'like']].mean())
dayorder = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
redcomm_daily['day'] = pd.Categorical(redcomm_daily['day'], categories=dayorder, ordered=True)
engagedaily = engagedaily.sort_values(by='day')
engagedaily
```

ENGAGEMENT BERDASARKAN PERIOD

```python
redcomm_period = redcomm[['period', 'like', 'comments', 'saved', 'share', 'engagement', 'video_views', 'reach', 'impression']]
redcomm_period
```

```python
engageperiod = pd.DataFrame(redcomm_period.groupby('period')[['engagement', 'reach', 'impression', 'like']].mean())
engageperiod
```
